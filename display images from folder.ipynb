```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

folder_path = r"C:\Users\Student\Desktop\images"

# Initialize an empty list to store the image data
image_data_list = []

# List all files in the folder
file_list = os.listdir(folder_path)
print("The files listed in the folder are:\n\n", file_list, "\n")

# Loop through each file in the folder
for filename in file_list:
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
        # Construct the full path to the image file
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)
        if image is not None:
            image_resized = cv2.resize(image, (640, 480))  # Resize all images to a common resolution
            image_data_list.append(image_resized)

print("Number of images read:\t", len(image_data_list), "\n")

# Converting the list data into image data array
image_data_array = np.array(image_data_list)
print("Shape of the image data array:", image_data_array.shape)

# Now plotting all the images one by one using matplotlib
num_images = image_data_array.shape[0]
rows = 1  # Adjust the number of rows and columns based on your preference
cols = num_images // rows + (1 if num_images % rows > 0 else 0)

plt.figure(figsize=(15, 10))
for i in range(num_images):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(cv2.cvtColor(image_data_array[i], cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    plt.title('Image {}'.format(i + 1))
    plt.axis('off')

plt.tight_layout()
plt.show()
```
